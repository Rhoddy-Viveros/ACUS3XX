Se da una distribución triangular de fuentes puntuales en campo libre. 
El nivel de presión sonora de las tres señales es de 75 dB a 1 m de distancia de cada fuente. 
Se coloca un punto receptor, R, a 3 m de cada fuente.

## Calcule el nivel de presión sonora en el punto de recepción para las siguientes señales:

### a) Tres señales identicas

```python
n = 3 # numero de fuentes
d = 3 # distancia en metros
d_ref = 1 # distancia de referencia en metros
L_ref = 75 # nivel de presion sonora de referencia en dB
```

#### Atenuación por distancia
$
L = 20*log_{10}(\frac{\tilde{\rho}}{\tilde{\rho_{0}}}) + 20*log_{10}( \frac{d_{ref}}{d})\\
L = L_{ref} + 20*log_{10}( \frac{d_{ref}}{d})\\
$
ó 
$
L = 20*log_{10}(\frac{\tilde{\rho}}{\tilde{\rho_{0}}} * \frac{d_{ref}}{d})\\
$

```python
L = L_ref + 20*np.log10(d_ref/d)
print('L = {0:.{1}f} dB'.format(L,1))
```

Si hay $n$ fuentes identicas a la misma distancia, la ecuación cambia a:
$
L = 20*log_{10}(\frac{\tilde{\rho}}{\tilde{\rho_{0}}} * \frac{d_{ref}}{d}*n)\\
L = L_{ref} + 20*log_{10}( n*\frac{d_{ref}}{d})\\
$
Entonces, en este caso:

```python
L3 = L_ref + 20*np.log10(n*d_ref/d)
print('L = {0:.{1}f} dB'.format(L3,1))
```

### b) Tres señales, dos identicas y una con desfase completo

```python
fs = 44100
A = 5
f = 15
w = 2*np.pi*f
t = np.linspace(0,1,fs)
sin1 = A*np.sin(w*t)
sin2 = A*np.sin(w*t+np.pi)
sin3 = np.around(sin1+sin2)
```

```python
fig,ax = plt.subplots(ncols=1,nrows=4,figsize=(6,8),
                      tight_layout=True,sharex=True)

ax[0].plot(t,sin1,'g')
ax[3].set_xlabel('Tiempo [s]')
ax[0].set_ylabel('Amplitud')
ax[1].set_ylabel('Amplitud')
ax[2].set_ylabel('Amplitud')
ax[3].set_ylabel('Amplitud')
ax[1].plot(t,sin2,'r')

ax[2].plot(t,sin1,'g')
ax[2].plot(t,sin2,'r')

ax[3].plot(t,sin3,'b')
ax[3].set_ylim([-A,A]);
```

Una señal de las identicas y la con la fase invertida son sumadas destructivamente, quedando una unica señal, por lo que su nivel de presión sonora es:
$
L = L_{ref} + 20*log_{10}( \frac{d_{ref}}{d})\\
$

```python
L = L_ref + 20*np.log10(d_ref/d)
print('L = {0:.{1}f} dB'.format(L,1))
```

### c) Tres señales, dos identicas y una con desfase parcial

```python
n_c = 2 # numero de fuentes identicas
L_c1 = L_ref + 20*np.log10(n_c*d_ref/d)
print('L = {0:.{1}f} dB'.format(L_c1,1))
```
