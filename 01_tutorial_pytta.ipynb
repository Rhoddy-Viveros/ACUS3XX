# Tutorial pyTTA

The [pyTTA](https://github.com/PyTTAmaster/PyTTa) library is a powerful tool for audio signal processing in Python, inspired by the [ITA](https://www.ita-toolbox.org) toolbox for MATLAB. It provides a wide range of functions for common audio processing tasks, such as measuring sound pressure levels, calculating room impulse responses, and analyzing sound spectra.

To install this library, simply run the following command: 

```bash 
pip install pytta
```

## Exploring pyTTA

Once you have installed this library, you will need to import it to use its functionalities.

````python
import pytta
````

## Read audiofile

````python
filename='audios/DemoITA.wav'
````

````python
signal = pytta.read_wav(filename)
````

## Metadata

````python
fs = signal.samplingRate
len_sig = signal.numSamples
time_len = round(signal.timeLength,3)
freqMin = signal.freqMin
freqMax = signal.freqMax
max_level = signal.max_level()[0]
rms = round(signal.rms()[0],3)
````

````python
print('-------INFO----------')
print('Sampling Rate: {}Hz'.format(fs))
print('Number of samples: {}'.format(len_sig))
print('Time Length: {}sec'.format(time_len))
print('Minimum Frequency: {}Hz'.format(freqMin))
print('Maximum Frequency: {}Hz'.format(freqMax))
print('Maximum level: {}dB'.format(max_level))
print('Root Mean Square value: {}'.format(rms))
````

## Plotting the signal

### Time domain

The variables `sig_timearray` and `time` correspond to the signal's amplitude information and temporal vector, respectively.

````python
sig_timearray = signal.timeSignal
time = signal.timeVector
````

````python
signal.plot_time(decimalSep='.');
````

````python
signal.plot_time_dB(decimalSep='.');
````

### Frequency domain

The variables `sig_freqarray` and `freq` correspond to the signal's magnitude information and frequential vector, respectively.

````python
sig_freqarray = signal.freqSignal
freq = signal.freqVector
````

````python
signal.plot_freq(decimalSep='.');
````

### Spectrogram

````python
signal.plot_spectrogram(decimalSep='.');
````
